```python
import requests
from consts import SESSION, ENDPOINT, SUMMARY_INDEX
import datetime
import json
```

```python
query = json.load(open("./facility_query.json", "r"))

response = SESSION.post(f"{ENDPOINT}/{SUMMARY_INDEX}/_search", data=json.dumps(query))
response_json = response.json()
response_json
```

```python
facilities = [v['key'] for v in response_json['aggregations']['facilities']['buckets']]
facilities = set(facilities)
```

```python
rg = requests.get("https://topology.opensciencegrid.org/api/resource_group_summary").json()
site_to_facility = { v['Site']['Name']: v['Facility']['Name'] for k,v in rg.items() }
print(site_to_facility)
```

```python
sites_osg_schedd = requests.get("https://raw.githubusercontent.com/osg-htc/ospool-data/master/data/active_site_report/latest.json").json()
facilities_osg_schedd = set([site_to_facility.get(v, None) for v in sites_osg_schedd])
```

```python
print(f"GRACC - OSG Schedd\n\n{facilities.difference(facilities_osg_schedd)}\n")
print(f"OSG Schedd - GRACC\n\n{facilities_osg_schedd.difference(facilities)}")
```
